```python
import pandas as pd
df=pd.read_csv('C:\\Users\\awezs\\downloads\\titanic_train.csv')
df.dropna()
```

```python
survival_count=df['Survived'].value_counts().rename(index={0:'Not Survived',1:'Survived'})
print(survival_count)
```

```python
survival_by_gender=df.groupby(['Survived','Sex']).size().rename(index={0:'Not Survived',1:'Survived'})
print(survival_by_gender)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x='Survived', hue='Sex', data=df, palette='cubehelix')
plt.show()
```

```python
survival_by_age=df.groupby(['Survived','Age']).size()
print(survival_by_age)
```

```python
plt.figure(figsize=(12,7))
sns.lineplot(x='Age',y='Survived',data=df, color='purple')
plt.xticks(range(0, 81, 10))
plt.show()
```

```python
passenger_class_counts = df['Pclass'].value_counts()

plt.figure(figsize=(12, 7))
sns.barplot(x=passenger_class_counts.index, y=passenger_class_counts.values, palette='Set2')
plt.show()
```
